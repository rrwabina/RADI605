## A. Data Preprocessing

Let's load the original dataset into a dataframe using <code>pandas</code>. The original dataset has 15,718 samples with 46 columns. It is vital to check its properties, such as the data types and null values, to preprocess the data prior data modeling. Results show that the dataset consists of null values that needs to be imputed. Several columns, however, contains too much null values that imputation may not necessarily and sufficiently work. 

```python
data = pd.read_csv('../data/Breast_Cancer.csv')
print(f'Original dataset: \t{data.shape}')
```

We created a function that determines columns that can be utilized for this probllem. While there's no concrete threshold in eliminating columns based on the number of their null values, most studies use 10%. This means that columns who have null values **less than 10%** of the samples will be retained in the dataset, otherwise removed. Results show that there are 36 features that remain in the dataset, such as <code>severity, bdate, marital, and etc.</code>

```python
def getcolumns_rate(dataframe, rate = 0.10): 
  less_than_50pct_nonnull = (dataframe.isnull().sum() / dataframe.shape[0]) < rate
  return dataframe.columns[less_than_50pct_nonnull]

remove_columns = getcolumns_rate(data)
print(remove_columns)
data = data[remove_columns]

data['bdate'] = data['bdate'].str[-4:]
data['age']   = 2565 - data['bdate'].astype('int')
data = data.drop('bdate', axis = 1)
```

We need to determine the categorical features in the dataset. These features should be encoded into numerical values because machine learning models expect features to be either floats or integers. Using the function <code>get_objectColumns</code>, we determined the columns that needs to be converted into numerical values. Some of these features include <code>severity, marital, breastfa, and etc.</code>

```python
def get_objectColumns(dataframe, type = 'object'):
    return list(dataframe.select_dtypes(include = type).columns)

encode_columns = get_objectColumns(data)
print(encode_columns)
for column in encode_columns:
    data[column] = LabelEncoder().fit_transform(data[column])

data['noova'] = data['noova'].replace([9.0], [3.0], inplace = False)
encode_columns = ['nobreast', 'nosecon', 'noova', 'inj', 'noparity']
for column in encode_columns:
    data[column] = LabelEncoder().fit_transform(data[column])
```

We used two techniques to impute the datasets - K-Nearest Neighbors (KNN) and Median. We adopted median than mean since median is more robust and it can mitigate the effect of outliers. The features <code>height</code> and <code>width</code> uses median imputation since it follows a non-parametric probability distribution. The remaining features <code>agemen</code> and <code>dur_brefed</code> are imputed through KNN since studies show that KNN also work with numerical and categorical features.

```python
def impute_knn(df, column_name, n_neighbors = 5):
    df_imputed = df.copy()
    imputer = KNNImputer(n_neighbors=n_neighbors)
    imputer.fit(df_imputed[[column_name]])
    df_imputed[column_name] = imputer.transform(df_imputed[[column_name]])
    return df_imputed

def impute_average(df, columns, impute_type = 'median'):
    df_imputed = df.copy()
    for col in columns:
        if impute_type == 'mean':
            impute_val = df_imputed[col].mean()
        elif impute_type == 'median':
            impute_val = df_imputed[col].median()
        else:
            raise ValueError('Invalid imputation type')
        df_imputed[col].fillna(impute_val, inplace=True)
    return df_imputed


data = impute_knn(data, 'agemen')
data = impute_knn(data, 'dur_brefed')
data = impute_average(data, ['weight'], 'median')
data = impute_average(data, ['height'], 'median')
```

## Ensemble Models: AdaBoost, Gradient Boosting, and XGBoost

Let's create the function <code>model_train</code> to simulate the model training and testing. The function returns nothing but only displays the model performance - including the accuracy score, confusion matrix, classification report, and the prediction time. 

```python
def model_train(model, X_train, y_train, X_test, y_test):
    start = time()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    class_names = ['Cancer', 'No Cancer']
    
    confusion = confusion_matrix(y_test, y_pred)
    
    print("Accuracy Score: ", np.round(accuracy_score(y_test, y_pred), 3))
    print('Confusion Matrix : \n', confusion)
    
    print(classification_report(y_test, y_pred, target_names = class_names))
    print(f"\t Fit and predict time: {np.round(time() - start, 3)} seconds")
```

### Adaptive Boosting

We adopted a decision tree-based Adaptive Boosting in this function called <code>train_adaboost</code>. This function requires the training set, testing set, and a boolean response if the user needs to adopt SMOTE. The default parameter utilized a class weight for any imbalanced classification when training the AdaBoost, otherwise it adopts SMOTE. In addition, this function uses Grid Search for hyperparameter tuning. 

```python
def train_adaboost(X_train, y_train, X_test, y_test, use_smote = False):
    if use_smote is True:
        smote = SMOTE(sampling_strategy = 'minority', k_neighbors = 10, random_state = 42)
        X_train, y_train = smote.fit_resample(X_train, y_train)

    param_grid = {'max_features': ['auto', 'sqrt', 'log2'],
                  'ccp_alpha'   : np.arange(0.001, 0.1, 0.01),
                  'max_depth'   : np.arange(1, 5),
                  'criterion'   : ['gini', 'entropy']}

    tree_class = DecisionTreeClassifier(max_depth = 5, random_state = 1024)
    grid_search = GridSearchCV(estimator = tree_class, 
                            param_grid = param_grid, 
                            cv = 5, 
                            verbose = True,
                            scoring = 'accuracy')

    grid_search = grid_search.fit(X_train, y_train)
    adaboost_dct = AdaBoostClassifier(grid_search.best_estimator_, 
                                    n_estimators  = 10, 
                                    random_state  = 42, 
                                    learning_rate = 0.001)
    model_train(adaboost_dct, X_train, y_train, X_test, y_test)
```

#### AdaBoost + GridSearchCV
The accuracy score on the test data was 0.994, indicating that the model performed very well in predicting the correct class for each data point. This model, however, is trained on an imbalanced datatset. The accuracy score is high, yet unbiased, since the model's performance is excellent for predicting 'No Cancer' class but poor performance for predicting 'Cancer' class. 

The precision score for 'No Cancer' class is 0.99, meaning that almost all predicted instances of 'No Cancer' were actually 'No Cancer'. The recall score for 'No Cancer' is 1.00, indicating that all actual instances of 'No Cancer' were correctly identified by the model.

```python
train_adaboost(X_train, y_train, X_test, y_test, use_smote = False)
```

#### AdaBoost + GridSearchCV + SMOTE (Training set)

The macro-average and weighted-average F1-scores are higher for 'Cancer' than in the previous model, indicating that the model's performance for predicting 'Cancer' has improved. However, it is still relatively poor compared to its performance for 'No Cancer'. The overall macro-average and weighted-average F1-scores are also higher than in the previous model, indicating that the model's overall performance has improved. This result may indicate the using SMOTE may provide significant improvements on the model's performance. 

```python
train_adaboost(X_train, y_train, X_test, y_test, use_smote = True)
```

### Gradient Boosting

Similar to AdaBoost, we adopted a decision tree-based Gradient Boosting on this function called <code>train_gradientboost</code>. This function requires the training set, testing set, and a boolean response if the user needs to adopt SMOTE. The default parameter utilized a class weight for any imbalanced classification when training the Gradient Boost, otherwise it adopts SMOTE. In addition, this function uses Grid Search for hyperparameter tuning. 

```python
def train_gradientboost(X_train, y_train, X_test, y_test, use_smote = False):
    if use_smote is True:
        smote = SMOTE(sampling_strategy = 'minority', k_neighbors = 10, random_state = 42)
        X_train, y_train = smote.fit_resample(X_train, y_train)
        
    param_grid = {'learning_rate': [0.1, 0.01, 0.001, 0.0001],
                'subsample': [1.0, 0.5],
                'max_features': np.arange(0, 10), 
                'max_depth': np.arange(0, 5)}

    gradient_boosting = GradientBoostingClassifier(n_estimators = 10, 
                                    learning_rate = 0.1, 
                                    random_state = 0)

    grid_search = GridSearchCV(estimator = gradient_boosting, 
                            param_grid = param_grid, 
                            cv = 5, 
                            verbose = True,
                            scoring = 'accuracy')
    model_train(grid_search, X_train, y_train, X_test, y_test)
```

#### Gradient Boosting + GridSearchCV

The model was fit using cross-validation with 5 folds and 400 candidate models were tried. The accuracy score on the test data was 0.993, indicating that the model performed very well in predicting the correct class for each data point. However, since it is trained on an imbalanced dataset, this may generate biased result to the 'No Cancer' class. This model (GB + GridSearchCV) was comparatively lower than AdaBoost + GridSearchCV. 

The AdaBoost model may have performed better because it is originally designed to handle imbalanced datasets by assigning greater weights to misclassified instances in the 'Cancer' class. Another factor is that AdaBoost trains a series of weak classifiers in sequence, with each subsequent classifier giving more weight to the misclassified instances from the previous classifier. This iterative process may improve the model's performance over time and can lead to better performance metrics overall. In contrast, Gradient Boosting trains a series of decision trees that are fit on the residuals of the previous tree. While this can lead to a more accurate model overall, it may not be as effective in handling imbalanced datasets or in focusing on correctly predicting the minority class.

```python
train_gradientboost(X_train, y_train, X_test, y_test, use_smote = False)
```

#### Gradient Boosting + GridSearchCV + SMOTE (Training set)

```python
train_gradientboost(X_train, y_train, X_test, y_test, use_smote = True)
```

### Extreme Gradient Boosting

```python
def train_xgboost(X_train, y_train, X_test, y_test, use_smote = False):
    if use_smote is True:
        smote = SMOTE(sampling_strategy = 'minority', k_neighbors = 10, random_state = 42)
        X_train, y_train = smote.fit_resample(X_train, y_train)
    xgboost = xgb.XGBClassifier(n_estimators = 10, 
                                objective  = 'binary:logistic',
                                reg_lambda = 0,
                                gamma  = 1,
                                max_depth = 6,
                                eta = 0.3,
                                scale_pos_weight = 10)
    xgboost.fit(X_train, y_train)
    model_train(xgboost, X_train, y_train, X_test, y_test)
```

#### XGBoost + GridSearchCV

Compared to the other models, this XGBoost model without SMOTE on an imbalanced dataset achieved a high accuracy score of 0.996. However, the recall score for the 'Cancer' class is relatively low at 0.41, indicating that the model is not able to identify all instances of the minority class. The XGBoost model without SMOTE seems to be the most accurate of the models presented, but its lower recall score for the minority class means that it may not be the best choice for identifying instances of that class. On the other hand, the AdaBoost and Gradient Boosting models were better at identifying the minority class, but at the expense of overall accuracy.

```python
train_xgboost(X_train, y_train, X_test, y_test, use_smote = False)
```

#### XGBoost + GridSearchCV + SMOTE

The XGBoost model with GridSearch and SMOTE has a slightly lower precision for detecting cancer than the XGBoost without SMOTE, but its recall is higher, meaning it is better at correctly identifying cancer cases. However, the accuracy score and other metrics are almost identical for both XGBoost models.

Compared to the other models, the XGBoost models with SMOTE and GridSearch have a higher accuracy score and better precision, recall, and F1-score for detecting cancer than the Gradient Boosting model with an imbalanced dataset. The AdaBoost model with GridSearchCV also has high accuracy and precision, but it has lower recall and F1-score for detecting cancer. Overall, the XGBoost model with SMOTE and GridSearchCV seems to be the best model for this imbalanced dataset based on the metrics reported.

```python
train_xgboost(X_train, y_train, X_test, y_test, use_smote = True)
```
